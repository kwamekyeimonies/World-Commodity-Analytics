```python
PreciousMetals=pd.read_excel("Precious_Metals.xlsx",sheet_name="MonthlyPrices")
PIndices=pd.read_excel("Precious_Metals.xlsx",sheet_name="MonthlyIndices")
```

```python
Gold_prices=PreciousMetals[["Time","Gold"]]
Gold_prices
```

```python
plt.figure(figsize=(20,10))
plt.plot(Gold_prices.Gold)
plt.xlabel("Time")
plt.ylabel("Gold")
plt.legend("Gold")
plt.show()
```

```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
CrudeOil_MPrices=pd.read_excel("CrudeOil.xlsx",sheet_name="MonthlyPrices")
CrudeOil_MIndices=pd.read_excel("CrudeOil.xlsx",sheet_name="MonthlyIndices")
```

```python
DubaiCrudeOil=CrudeOil_MPrices[['Time','CrudeOil_Dubai']]
DubaiCrudeOil
```

```python
plt.figure(figsize=(20,20))
plt.plot(CrudeOil_MPrices.CrudeOil_Dubai)
plt.xlabel("Time", size=10)
plt.ylabel("Prices of Dubai Crude Oil",size=10)
plt.legend(["Dubai Oil"])
plt.title("Dubaian Crude Oil Graph",size=20)
plt.grid(True)
plt.show()
```

```python
DubaiCrudeOil.describe()
```
